### Diccionario:

16. **Engine Type:** Tipo de motor del vehículo.
17. **Engine Size:** Tamaño del motor del vehículo.
18. **Engine Cylinder Count:** Número de cilindros en el motor.

20. **Heavy-Duty Power System:** Sistema de energía pesado para vehículos de trabajo pesado.
21. **Notes:** Notas adicionales o información sobre el vehículo.

Al profundizar en el análisis de los datos, se destaca la presencia de columnas con un alto número de valores faltantes, como PHEV Range, Transmission Make, Heavy-duty Power System, Notes y Number of Passengers. Estas columnas podrían no ser críticas para nuestro análisis, ya que la falta de información podría afectar la interpretación y utilidad de los resultados. La ausencia de datos en estas categorías podría deberse a varias razones, como la falta de disponibilidad de información o la naturaleza específica de ciertos vehículos que no requieren estas especificaciones.

En el caso de Engine Size y Engine Cylinder Count, se podría argumentar que ambas proporcionan información similar sobre las características del motor. Dado que Engine Size se considera una medida más integral del motor, podría ser preferible mantener esta variable y descartar Engine Cylinder Count para simplificar el conjunto de datos y reducir la complejidad sin perder información esencial.

Asimismo, es importante considerar que el análisis de datos se centra en aspectos relevantes para la producción de CO2, y la relación entre Engine Size y las emisiones de CO2 podría ser un área clave de exploración. Esto destaca la necesidad de realizar elecciones estratégicas en cuanto a las variables a incluir en el análisis para obtener resultados más significativos y aplicables a la problemática abordada.

```python
df1_filtered = df1[['Category', 'Model', 'Model Year', 'Manufacturer', 'Fuel', 'All-Electric Range','Alternative Fuel Economy Combined','Conventional Fuel Economy Combined', 'Transmission Type','Engine Size']]
```

```python
df1_filtered.head(3)
```

```python
df1_filtered['Category'].unique()
```

```python
df1_filtered['Model'].unique()
```

```python
df1_filtered['Model Year'].unique()
```

```python
df1_filtered['Model Year'] = df1_filtered['Model Year'].apply(lambda x: int(x) if pd.notnull(x) else int(0))
```

```python
df1_filtered['Model Year'].unique()
```

```python
df1_filtered['Manufacturer'].unique()
```

```python
df1_filtered['Transmission Type'].unique()
```

```python
df1_filtered['Transmission Type'] = df1_filtered['Transmission Type'].apply(lambda x : 'Automatic|Manual' if x == 'Manual|Automatic' else x)
df1_filtered['Transmission Type'] = df1_filtered['Transmission Type'].apply(lambda x: 'Manual' if pd.isnull(x) else x)
df1_filtered['Transmission Type'] = df1_filtered['Transmission Type'].replace({'Auto': 'Automatic'})
```

```python
df1_filtered['Transmission Type'].unique()
```

```python
df1_filtered['Transmission Type'].value_counts()
```

Además, consideramos la relación entre Engine Size y Engine Cylinder Count, optando por conservar Engine Size debido a su naturaleza más integral y, posiblemente, más relevante en términos de consumo de combustible y emisiones de CO2. Este enfoque de refinamiento busca simplificar el conjunto de datos sin perder información crítica, permitiéndonos concentrarnos en variables clave que influyen en nuestro análisis.

Tras estas consideraciones, procedimos a eliminar posibles duplicados en nuestro conjunto de datos filtrado df1_filtered.
